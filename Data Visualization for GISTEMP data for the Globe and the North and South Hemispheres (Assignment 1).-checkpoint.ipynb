```python
# Directory paths to the yearly and monthly data files for visualization
yearly_data = pd.read_csv("C:/Users/ZUZILE/Downloads/Video/Data Visualization/Week 2/Assignment 1/Programming_Assignment_1_Data_New/ExcelFormattedGISTEMPData2CSV.csv")
monthly_data = pd.read_csv('C:/Users/ZUZILE/Downloads/Video/Data Visualization/Week 2/Assignment 1/Programming_Assignment_1_Data_New/ExcelFormattedGISTEMPDataCSV.csv')
```

```python
### Here we plot the data of global temperature in relation with the frequency using the Bar graph
plt.figure(figsize=(10,6.18));
yearly_data['Glob'].plot.hist(alpha=0.5,bins=15)
plt.xlabel('Global Temperature (Celsius)')
```

```python
# Displaying data using grid in order to be able view the decline/uprise of the temperature
plt.figure(figsize=(10,6.18));
plt.plot(yearly_data['Glob'])
plt.xlabel('Global Temperature (Celsius)')
plt.ylabel('Deviation')
plt.grid(b=True, color='aqua', alpha=0.3, linestyle='-.', linewidth=2)
```

```python
# Ploting the bar graphs of the global temperature, nothern hemisphere, and southern hemisphere for comparison
plt.figure(figsize=(10,6.18));
plt.figure();
yearly_data['Glob'].plot.hist(alpha=0.5)
yearly_data['NHem'].plot.hist(alpha=0.5)
yearly_data['SHem'].plot.hist(alpha=0.5)
plt.xlabel('Global Temperature (Celsius)')
```

```python
# Using seaborn module to colorfuly display the above data visualization
plt.figure(figsize=(40,10));
sn.barplot(x = yearly_data['Glob'], y = yearly_data['Year'])
plt.xlabel('Global Temperature (Celsius)')
```

```python
plt.figure();
scatter_matrix(yearly_data, alpha=0.5, figsize=(36, 36), diagonal='kde')
plt.show()
```

```python
plt.figure(figsize=(10,6.18));
plt.plot(yearly_data['Year'], yearly_data['Glob'])
plt.plot(yearly_data['Year'], yearly_data['NHem'])
plt.plot(yearly_data['Year'], yearly_data['SHem'])
plt.legend(['Global Temperature','Northen Hemisphere','Southen Hemisphere'])
plt.xlabel('Year')
plt.ylabel('Average Temperature (Celsius)')
plt.grid(b=True, color='red', alpha=0.3, linestyle='-.', linewidth=2)
```

```python
yearly_data.plot.scatter(x='NHem', y='SHem',figsize=(10,6.18))
plt.xlabel('Northern Hemisphere (Celsius)')
plt.ylabel('Southern Hemisphere (Celsius)')
plt.title('Temperature Correlation of Two Hemispheres (1880-2015)')
plt.grid(b=True, color='red', alpha=0.3, linestyle='-.', linewidth=2)
plt.show()
```

## Monthly data visualization.

```python
monthly_data.columns
```

```python
# Displaying the first 5 columns for the monthly data
monthly_data.head()
```

```python
plt.figure();
scatter_matrix(monthly_data, alpha=0.9, figsize=(36, 36), diagonal='kde')
plt.show()
```

```python
# Clearing incorrect data
months = ['Year', 'Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep','Oct', 'Nov', 'Dec']
monthly_data = monthly_data.replace(['***','****'], ['NaN','NaN'])[months]
```

```python
ax = plt.figure(figsize=(10,6.18))
ax = parallel_coordinates(monthly_data[monthly_data['Year']>2010],'Year')
plt.xlabel('Months (per year)')
ax.legend().remove()
```
